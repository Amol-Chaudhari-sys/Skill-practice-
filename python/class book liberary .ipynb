```python
class lib:
    def __init__(self):
        self.no_of_book=0
        self.name=[]

    def addbook(self,name):
        self.name.append(name)
        self.no_of_book+=1
        
    def show(self):
        print (f"no of books are {self.no_of_book }  or {len(self.name)}" )
        for book in self.name:
            print(book) 


```

```python
l1=lib()
l1.addbook("harry porter 1")
l1.addbook("harry porter 2")
l1.addbook("harry porter 3")
l1.addbook("harry porter 4")
l1.addbook("harry porter 5")
l1.addbook("harry porter 6")
```

```python
print (lib.show(l1))
```
