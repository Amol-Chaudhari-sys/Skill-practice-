```python
def fact(n):
    if n ==0 or n==1:
        return 1
    return (n*fact(n-1))
```

```python
fact(5)
```

```python
def fibo (n):
    if n<=0 :
        return 0 
    elif n==0:
        return 1 
    elif (n==2):
        return 1
    else :
        return fibo (n-1)+fibo (n-2)
    
```

```python
fibo(7)
```

```python
def fibo (n):
    series=[0,1]

    for _ in range (2,n ):
        series.append(series [-1]+series[-2])
    return series
```

```python
fibo (8)
```
