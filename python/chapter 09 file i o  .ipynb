```python
import random 
def game ():
    score = random.randint(1,50)
    print (f"your score is {score }")
    with open ("highscore.txt") as f:
        hscore=f.read()
        if (hscore != ""):
            hscore= int (hscore)
    
        else :
            hscore = 0
        if (score >hscore):
            with open ("highscore.txt",'w') as f:
                f.write (str(score))
        return score        
            


    
```
