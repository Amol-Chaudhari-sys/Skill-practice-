```python
st= input("Enter the world ")
coding = (input ("Enter the 1 for Encoding 0 for decoding "))
words = st.split(" ")
nst=[]
coding =True if (coding =="1") else False 
if coding :
    print ("Waiting for decoding" )
    for word in words:
        print (word)
        if (len(word)>=3):
            r1="uyw"
            r2="vbn"
            nwd=r1+word[1:]+word[0]+r2
            nst.append(nwd)
        else :
            nst.word[::-1]
    print (" ".join(nst))
else :
    print ("decoding ")
    for word in words:
        print (word)
        if (len(word)>=3):
            nwd=word[-1]+word[3:-3]
            nst.append(nwd)
            
        else :
            nst.word[::-1]
    print (" ".join(nst))
            
            


            

    
```
