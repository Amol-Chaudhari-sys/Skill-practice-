# z-Test

```python
import scipy.stats as st
import numpy as np 

p_u=82
p_std=20
s_x=90
no_s=81
ap=0.05

z_cal= (s_x-p_u)/(p_std/np.sqrt(no_s))
print (z_cal)
z_table=st.norm.ppf(1-ap)
print(z_table)
```

```python
if z_table <z_cal:
    print ("ha is right "+ str(z_cal))
else :
    print ("H0 is right ")
```

```python
# u old = u new 
# u = old - new 
# x = old -new 
old =np.array([45,42,38,43,41,44,40,42,39,41,44,39,42,40,43,42,41,39,44,43,38,44,43,39,37,44,45,40,42,41])
new =np.array([48,49,50,47,48,49,48,50,49,49,49,48,48,49,50,49,50,48,48,49,50,48,47,45,50,49,47,47,49,50])
```

```python
print (new)
```

```python
old_x=np.mean(old)
new_x=np.mean (new)
pop_std=2.5
noofsample=len(new)
```

```python
z_cal= (new_x-old_x)/(pop_std/np.sqrt(noofsample))
z_table = st.norm.ppf(0.95)
```

```python
if z_table <z_cal:
    print ("ha is right "+ str(z_cal))
else :
    print ("H0 is right ")
```

# T-Test

```python
T_test=( x-u)/(s/sqrt n) # s= std of sample
```

```python
u =150
x=148
n=25
s=5 
a=0.05

t_cal=(u-x)/(5/np.sqrt(n))
```

```python
t_table=st.t.ppf(a,n-1)  # degree of r=freedome =n-1
t_table
```

```python
if t_table <t_cal:
    print ("ha is right "+ str(t_cal))
else :
    print ("H0 is right ")
```

```python
row1=np.array([40,45,25,10])
row2=np.array([35,30,20,30])

sum1=np.sum(row1)
sum2 = np.sum(row2)
sumrow= np.array([sum1,sum2])
sumcolmn=row1+row2
sumcolmn
```

```python
exp=[]
for i in sumrow:
    for j in sumcolmn:
        v=(i*j)/235
        exp.append(v)
exp
```

```python
ob=np.array([40,45,25,10,35,30,20,30])
```

```python
np.sum(np.square(ob-exp)/exp)
```
