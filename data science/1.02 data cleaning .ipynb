```python
# vbackword filling 
data3.fillna(method="bfill",axis=1)
```

```python
# forword filing 
data3.fillna(method="ffill",axis =1)
```

```python
data3["historical_default"].fillna(data3["historical_default"].mode()[0],inplace=True)
```

```python
data3 
```

```python
for i in data3.select_dtypes(include="object").columns:
    data3[i].fillna(data3[i].mode()[0],inplace=True)
```
